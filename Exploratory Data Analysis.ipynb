```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Train_data.csv')

# Print the columns of the DataFrame to check their names
print("Columns in DataFrame:", df.columns)

# Specify the columns you want to convert to numeric
columns_to_convert = ['Classification', 'Reported_Value']

# Convert the specified columns to numeric
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Now the specified columns have been converted to numeric
# You can check the data types of the columns using df.dtypes
print("Data types after conversion:", df.dtypes)

# Now you can calculate the correlation
correlation = df['Classification'].corr(df['Reported_Value'])

# Print the correlation value
print(f"Correlation between Classification and Reported Value: {correlation}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Train_data.csv')

# Create a contingency table (cross-tabulation)
contingency_table = pd.crosstab(df['Classification'], df['Reported_Value'])

# Create a heatmap
plt.figure(figsize=(12, 5))
sns.heatmap(contingency_table, cmap='coolwarm', annot=True, fmt="d", linewidths=.5)

# Set plot labels and title
plt.xlabel('Reported_Value')
plt.ylabel('Classification')
plt.title('Correlation between Classification and Reported_Value')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
data = pd.read_csv('Train_data.csv')

# Display basic information about the dataset
print(data.info())
```

```python
# Display summary statistics for numerical columns
print(data.describe())
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
#Check for missing values
print(data.isnull().sum())
```

```python
# Distribution of Activity Type with increased x and y axis
plt.figure(figsize=(16, 7))  # Adjust the figsize as needed
sns.countplot(x='Activity Type', data=data)
plt.title('Distribution of Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Count')  # Add ylabel to the vertical axis
plt.show()
```

```python
# Distribution of Classification with increased x and y axis
plt.figure(figsize=(12, 8))  # Adjust the figsize as needed
sns.countplot(x='Classification', data=data)
plt.title('Distribution of Classification')
plt.xlabel('Classification')
plt.ylabel('Count')  # Add ylabel to the vertical axis
plt.show()
```

```python
# Box plot of Reported_Value by Activity Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='Activity Type', y='Reported_Value', data=data)
plt.title('Box plot of Reported_Value by Activity Type')
plt.show()
```

```python
# Scatter plot of Reported_Value by Activity Source
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Activity Source', y='Reported_Value', data=data)
plt.title('Scatter plot of Reported_Value by Activity Source')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair plot for numerical variables with increased size
sns.pairplot(data[['Reported_Value']], diag_kind='kde', size=2.5, aspect=1.2)
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```
